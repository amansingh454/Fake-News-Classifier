```python
import re
import nltk
```

```python
from nltk.corpus import stopwords
import string 
string.punctuation
import heapq
from nltk.stem import PorterStemmer
from nltk import WordNetLemmatizer
lm=WordNetLemmatizer()
ps=PorterStemmer()
```

```python
data['title'].head()
```

```python
for i in range(len(data)):
    title=data.iloc[i,0:1][0]
    title=re.sub(r"\W"," ",title)
    title=re.sub(r"\d"," ",title)
    title=re.sub(r"\s+"," ",title)
    title=title.lower()
    data.iloc[i,0:1]=title
```

```python
title_corpus=[]
for i in range(len(data)):
    title=data.iloc[i,0:1][0]
    words=nltk.word_tokenize(title)
    words=[word for word in words if len(word)>1]
    words=[word for word in words if word not in stopwords.words('english')]
    words=[lm.lemmatize(word) for word in words]
    words=" ".join(words)
    title_corpus.append(words)
    data.iloc[i,0:1]=words
```

```python
for i in range(len(data)):
    text=data.iloc[i,1:2][0]
    text=re.sub(r"\W"," ",text)
    text=re.sub(r"\d"," ",text)
    text=re.sub(r"\s+"," ",text)
    text=text.lower()
    data.iloc[i,1:2]=text
```

```python
for i in range(len(data)):
    text=data.iloc[i,1:2][0]
    words=nltk.word_tokenize(text)
    words=[word for word in words if len(word)>1]
    data.iloc[i,1:2]=" ".join(words)
```

```python
text_corpus=[]
for i in range(len(data)):
    text=data.iloc[i,1:2][0]
    words=nltk.word_tokenize(text)
    words=[lm.lemmatize(word) for word in words]
    words=" ".join(words)
    text_corpus.append(words)
    data.iloc[i,1:2]=words
    
```

```python
data.reset_index(inplace=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
tf=TfidfVectorizer(max_features=10000,ngram_range=(1,3))
```

```python
title_vector=tf.fit_transform(title_corpus).toarray()
text_vector=tf.fit_transform(text_corpus).toarray()
```

# working on title column

```python
X=title_vector
y=data['label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    See full source and example: 
    http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html
    
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

# Naive bayes

```python
from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB()
```

```python
from sklearn import metrics
import numpy as np
import itertools
```

```python
classifier.fit(X_train, y_train)
pred = classifier.predict(X_test)
score = metrics.accuracy_score(y_test, pred)
print("accuracy:   %0.3f" % score)
cm = metrics.confusion_matrix(y_test, pred)
plot_confusion_matrix(cm, classes=['FAKE', 'REAL'])
```

# Passive Aggressive Classifier

```python
from sklearn.linear_model import PassiveAggressiveClassifier
linear_clf = PassiveAggressiveClassifier()
```

```python
linear_clf.fit(X_train, y_train)
pred = linear_clf.predict(X_test)
score = metrics.accuracy_score(y_test, pred)
print("accuracy:   %0.3f" % score)
cm = metrics.confusion_matrix(y_test, pred)
plot_confusion_matrix(cm, classes=['FAKE Data', 'REAL Data'])
```

# Multinomial classifier with parameter

```python
classifier=MultinomialNB(alpha=0.1)
```

```python
previous_score=0
for alpha in np.arange(0,1,0.1):
    sub_classifier=MultinomialNB(alpha=alpha)
    sub_classifier.fit(X_train,y_train)
    y_pred=sub_classifier.predict(X_test)
    score = metrics.accuracy_score(y_test, y_pred)
    if score>previous_score:
        classifier=sub_classifier
    print("Alpha: {}, Score : {}".format(alpha,score))
```
